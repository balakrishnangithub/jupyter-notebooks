```
import pandas as pd
import altair as alt
# Load Data
df = pd.read_csv('https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
# Remove unwanted columns
df = df.drop(columns=['Province/State', 'Lat', 'Long'])
# Sum Province/State level count to be Country level count
df = df.groupby(['Country/Region']).sum().reset_index()
# Prepare data for chart
df = df.melt(id_vars='Country/Region')
df['variable'] = pd.DatetimeIndex(df['variable'])
# Render Chart
alt.Chart(df[df['Country/Region'].isin(
    ['US','Spain','Italy','Germany','China','India']
    )]).mark_line().encode(
        x='variable', y='value', color='Country/Region'
        ).interactive()
```
